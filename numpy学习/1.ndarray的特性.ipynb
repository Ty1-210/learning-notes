```python
import numpy as np
```

多维性

```python
arr = np.array(5) #创建0维ndarray
print(arr)
print('arr的维度:' ,arr.ndim)
```

```python
arr = np.array([1, 2, 3])  
print(arr)
print('arr的维度:', arr.ndim)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]]) 
print(arr)
print('arr的维度:', arr.ndim)
```

同质性

```python
arr = np.array([1, 'hello']) #不同数据类型会被强制转换成相同类型
print(arr)
```

```python
arr = np.array([1, 2.5]) #1变成了浮点数
print(arr)
```
