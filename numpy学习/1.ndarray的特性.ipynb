## ndarray 的多维性

NumPy 数组的核心是**多维同质数据容器**。和 Python 列表的关键区别：

| | list | ndarray |
|---|---|---|
| 维度 | 嵌套实现，不规则 | 真正的 N 维张量 |
| 存储 | 存指针 + 散落对象 | 连续内存块 |
| 操作 | 循环逐元素 | 向量化，一条指令操作全数组 |

```python
np.array([1, 2, 3])         # 1-D (3,)
np.array([[1,2],[3,4]])     # 2-D (2, 2)
np.array([[[1]]])           # 3-D (1, 1, 1)
```

> **核心优势**：向量化运算——不用写循环，`a + b` 就把整个数组逐元素加了。速度快几个数量级。


````python
arr = np.array(5) #创建0维ndarray
print(arr)
print('arr的维度:' ,arr.ndim)
````

````python
arr = np.array([1, 2, 3])  
print(arr)
print('arr的维度:', arr.ndim)
````

````python
arr = np.array([[1, 2, 3], [4, 5, 6]]) 
print(arr)
print('arr的维度:', arr.ndim)
````

## ndarray 的同质性

一个 ndarray 里**所有元素的类型相同**（由 `dtype` 决定）。

```python
a = np.array([1, 2.5, 3])      # 混合输入 → 自动提升为 float64
a = np.array([1, 2, 'hello'])  # 全部变成字符串 U21
```

**好处**：
- 连续内存，CPU 可以批量读取，远快于 Python list
- 每个元素占用相同字节数，不用存指针，内存利用率极高

**代价**：不能像 list 那样随便混装不同类型。鱼和熊掌。


````python
arr = np.array([1, 'hello']) #不同数据类型会被强制转换成相同类型
print(arr)
````

````python
arr = np.array([1, 2.5]) #1变成了浮点数
print(arr)
````
