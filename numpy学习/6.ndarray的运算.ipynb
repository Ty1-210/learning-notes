算术运算

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
print(a-b)
print(a*b)
print(a/b) #乘除法是对应位置运算 而非矩阵运算
print(a**2)
print(a**b)
```

```python
a = [1,2,3]
b = [4,5,6]
print(a+b) #list array运算规则不同
```

```python
t = np.array([[1,2,3],[4,5,6]])
d = np.array([[7,8,9],[10,11,12]])
print(t+d)
print(t-d)
print(t*d)
print(t/d)
print(t**2)
```

```python
print(t+3) #每个元素都和标量运算
```

广播机制规则 


1：屁股对齐（右对齐）

如果两个数组的维度数量不一样，先把它们的最右边（尾部）对齐。

比如：一个形状是 (4, 3)，另一个是 (3,)。就把它们右边的 3 对齐。


规则 2：缺啥补啥（往左边补 1）

对齐后，如果哪个数组的维度比较少（比较短），就在它的最左边补上数字 1，直到两个数组的维度长度一样为止。

比如：(3,) 比 (4, 3) 短，就在左边补个 1，变成 (1, 3)。


规则 3：检查能不能拉伸（兼容原则）

现在两个数组一样长了，我们从左到右逐个对比每一位的数字。只要满足下面任意一种情况，就是兼容的（可以广播）：

相等：两个数字一样（比如都是 3）。

其中一个是 1：只要有一个是 1，它就可以被无限拉伸去迁就另一个数字。

如果既不相等，也没有一个是 1（比如一个是 4，一个是 5），那就直接报错，没法算了！

```python
a = np.array([1,2,3])
b = np.array([[4],[5],[6]])
print(a+b)
print(a-b)
```

矩阵运算

```python
t = np.array([[1,2,3],[4,5,6]])
d = np.array([[7,8],[9,10],[11,12]])
print(t@d) #优先级与乘除法一致
```

```python
a = np.array([1,2,3])
b = np.array([[1,2,3],[4,5,6]])
c = np.array([[4],[5],[6]])
print(b@a) #a:3*1
print(a@c) #a:1*3
#NumPy不区分行列向量，会根据运算自动变换为行或列向量
```
