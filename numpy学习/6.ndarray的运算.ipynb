## 算术运算

NumPy 算术都是**逐元素**操作，不需要写循环。

```python
a + b      # 逐元素加
a - b      # 逐元素减
a * b      # 逐元素乘（不是矩阵乘法！）
a / b      # 逐元素除
a ** 2     # 逐元素平方
np.sqrt(a) # 逐元素开方
a @ b       # 矩阵乘法（Python 3.5+）
np.dot(a, b) # 也是矩阵乘法
```

> **`a * b` 是逐元素乘法，不是矩阵乘法！** 矩阵乘法用 `@` 或 `np.dot()`。这是 NumPy 新手的第一大坑。


````python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
print(a-b)
print(a*b)
print(a/b) #乘除法是对应位置运算 而非矩阵运算
print(a**2)
print(a**b)
````

````python
a = [1,2,3]
b = [4,5,6]
print(a+b) #list array运算规则不同
````

````python
t = np.array([[1,2,3],[4,5,6]])
d = np.array([[7,8,9],[10,11,12]])
print(t+d)
print(t-d)
print(t*d)
print(t/d)
print(t**2)
````

````python
print(t+3) #每个元素都和标量运算
````

## 广播机制（Broadcasting）

不同形状的数组做算术时，NumPy 会自动"拉伸"小的去匹配大的，条件是维度从**右边对齐**后，每个维度要么相等、要么其中一个为 1。

### 规则
1. 维度数不同时，左边补 1
2. 从右往左逐个维度比对：相等或有一个是 1 就能广播
3. 尺寸=1 的维度被拉伸到匹配对方

```python
a.shape = (3, 4)
b.shape = (4,)      # 右对齐：3 vs 无 → 补1, 4 vs 4 OK
# → b 被广播成 (1,4) → (3,4)，逐元素加

a.shape = (3, 1)
b.shape = (1, 4)
# → (3,1) + (1,4) 广播为 (3,4) + (3,4)
```

> 广播让你不用显式复制数据就能对不同形状的数组做运算——NumPy 做了"心算"，没有真拷贝。


````python
a = np.array([1,2,3])
b = np.array([[4],[5],[6]])
print(a+b)
print(a-b)
````

## 矩阵运算

```python
a @ b               # 矩阵乘法
np.dot(a, b)        # 同上
np.linalg.inv(a)    # 逆矩阵
np.linalg.det(a)    # 行列式
np.linalg.eig(a)    # 特征值和特征向量
a.T                 # 转置
a.reshape(2, 3)     # 改变形状（总元素数不变）
```


````python
t = np.array([[1,2,3],[4,5,6]])
d = np.array([[7,8],[9,10],[11,12]])
print(t@d) #优先级与乘除法一致
````

````python
a = np.array([1,2,3])
b = np.array([[1,2,3],[4,5,6]])
c = np.array([[4],[5],[6]])
print(b@a) #a:3*1
print(a@c) #a:1*3
#NumPy不区分行列向量，会根据运算自动变换为行或列向量
````
