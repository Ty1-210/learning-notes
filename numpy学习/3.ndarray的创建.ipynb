````python
import numpy as np
````

## 基础创建

```python
np.array([1, 2, 3])              # 从列表
np.array([[1,2],[3,4]])          # 从嵌套列表 → 2-D
np.array([1, 2, 3], dtype=float) # 指定类型
```

每次从 Python list 转 ndarray 都会拷贝数据。如果数据已经有 numpy 数组，考虑 `np.asarray()`（不拷贝）。


````python
data = [4, 5, 6]
arr1 = np.array(data)
print(arr1)
arr2 = np.array(data, dtype=np.float64) #dtype强制转换类型
print(arr2)

#copy
arr = np.copy(arr1) #arr的元素和参数里的数组相同，但不是一个数组
print(arr)
arr1 = [4, 5, 8]
print(arr)
````

## 预定义形状的创建

```python
np.zeros((3, 4))        # 全 0
np.ones((2, 3))         # 全 1
np.empty((2, 3))        # 分配内存但不初始化（速度最快，值是垃圾）
np.full((2, 3), 7)      # 全填 7
np.eye(3)               # 3×3 单位矩阵
np.identity(3)          # 同上
```

> `np.empty()` 不初始化，用于先分配后覆盖的场景（追求速度），不要直接用于计算。


````python
#全0
arr = np.zeros((2,3)) #定义两行三列的全零
print(arr)
print(arr.dtype)
arr = np.zeros((2,3), dtype=int) 
print(arr)
print(arr.dtype)
````

````python
arr = np.zeros((3,)) #少维度就不写
print(arr)
````

````python
#全1
arr = np.ones((2,3), dtype=int)
print(arr)
````

````python
#未初始化
arr = np.empty((4,5)) #随即填充
print(arr)
arr = np.empty((4,5), dtype=int)
print(arr)
````

````python
#指定其他数值
arr = np.full((2,3), 1234)
print(arr)
````

````python
#like
print(arr)
arr1 = np.zeros_like(arr) #构建形状相同的,可用dtype转换类型,否则照搬原本类型
print(arr1)
arr2 = np.ones_like(arr)
print(arr2)
arr3 = np.full_like(arr, 12)
print(arr3)
arr4 = np.empty_like(arr, dtype=float)
print(arr4)
````

## 基于数值范围

```python
np.arange(0, 10, 2)     # [0, 2, 4, 6, 8] — 类似 Python range
np.linspace(0, 1, 5)    # 等间距 5 个点：[0., 0.25, 0.5, 0.75, 1.]
np.logspace(0, 2, 5)    # 对数等间距：10^0 到 10^2
```

- `arange`：等差整数序列，给步长
- `linspace`：精确控制点数，**包含**终点（和 arange 不同）


````python
#等差数列
arr= np.arange(1,10,3) #start,end,step 包含start 不包含end
print(arr)
````

````python
#等间隔数列
arr = np.linspace(1,10,5) #start,end,num 包含end
print(arr)
arr = np.linspace(1,10,5, dtype=int) 
print(arr)
````

````python
#对数间隔数列
arr = np.logspace(0,4,3,base=2) #start,end,num 包含end 可用dtype,默认为浮点数,以10为底
#0到4分成三分,得0,2,4 取2为底的指数,得1,4,16
print(arr)
````

## 特殊矩阵生成

```python
np.diag([1, 2, 3])           # 对角矩阵
np.triu([[1,2],[3,4]])       # 上三角
np.tril([[1,2],[3,4]])       # 下三角
```


````python
#单位矩阵
arr = np.eye(4)
print(arr)
arr1 = np.eye(3,4,dtype=int)
print(arr1)
````

````python
#对角矩阵
arr = np.diag([1,2,3,5])
print(arr)
````

## 随机数组的生成

```python
np.random.random((3, 4))          # [0, 1) 均匀分布
np.random.rand(3, 4)              # 同上，写法更简洁
np.random.randn(3, 4)             # 标准正态分布 N(0,1)
np.random.randint(0, 10, (3, 4))  # [0, 10) 随机整数
np.random.normal(loc=0, scale=1, size=(3, 4))  # 指定均值和标准差的正态
np.random.uniform(low=0, high=1, size=(3, 4))   # 指定区间的均匀分布
```


````python
#0-1间均匀分布随机浮点数
arr = np.random.rand(2,3)
print(arr)
````

````python
#生成指定范围区间均匀分布随机浮点数
arr = np.random.uniform(3,6,(2,3)) #start,end,shape 包含start 不包含end
print(arr)
````

````python
#生成指定范围区间均匀分布随机整数
arr = np.random.randint(3,30,(2,3)) #start,end,shape 包含start 不包含end 一维数组则shape直接输入一个int即可
print(arr)
````

````python
#标准正太分布随机数组
arr = np.random.randn(2,3)
print(arr)
````

````python
#任意正态分布随机数组
arr = np.random.normal(10,1,10) #mean std size size也可以是元组 创建一个矩阵或高阶张量
print(arr)
````

## 设置随机种子

```python
np.random.seed(42)    # 固定随机数，保证每次运行结果一致
```

> 做实验和调试时必须设置种子——否则每次结果不同，没法复现和排错。42 是社区惯例数字。


````python
np.random.seed(20) #设置种子后随机数组生成就一致了
arr = np.random.randint(1,10,(2,5))
print(arr)
````
