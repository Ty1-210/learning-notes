索引从0开始

```python
import numpy as np
```

一维数组

```python
arr = np.random.randint(1,100,30)
print(arr)
```

```python
print(arr[3])
print(arr[:])
print(arr[2:6]) #start:end 含start 不含end
print(arr[(arr>10)&(arr<70)]) #布尔索引 支持& |
print(arr[slice(2,6)]) #slice中可写 start end step 不写step等价于2:6
```

二维数组

```python
arr = np.random.randint(1,100,(4,6))
print(arr)
```

```python
print(arr[0,3])
print(arr[:,3])
print(arr[1:3,3])
print(arr[arr>50]) #对二维数组用布尔索引返回一维数组

print(arr[2][arr[2]>50]) #第二行所有大于50的树
print(arr[:,3][arr[:,3]>50]) #第三列所有大于50的数
```
