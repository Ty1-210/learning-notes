## 索引从 0 开始

NumPy 索引规则和 Python list 一致：从 0 开始，含头不含尾。

```python
a[0]       # 第一个
a[-1]      # 最后一个
a[1:4]     # 索引 1,2,3（不含 4）
a[::-1]    # 反转数组
```

> NumPy 切片返回的是**视图（view）**而非拷贝——改切片会改原数组。和 list 切片完全不同。


````python
import numpy as np
````

## 一维数组

```python
a[2]         # 下标索引
a[1:5]       # 切片
a[[0, 2, 4]] # 花式索引（fancy indexing）— 跳着取
a[a > 0]     # 布尔索引 — 条件筛选
```

花式索引返回**拷贝**（不是视图），布尔索引也是拷贝。


````python
arr = np.random.randint(1,100,30)
print(arr)
````

````python
print(arr[3])
print(arr[:])
print(arr[2:6]) #start:end 含start 不含end
print(arr[(arr>10)&(arr<70)]) #布尔索引 支持& |
print(arr[slice(2,6)]) #slice中可写 start end step 不写step等价于2:6
````

## 二维数组

```python
a[0, 1]          # 第 0 行第 1 列
a[0]             # 第 0 行整行
a[:, 1]          # 第 1 列整列（: 表示"所有行"）
a[0:2, 1:3]      # 前两行、中间两列的子矩阵
a[[0, 2], :]     # 取第 0 行和第 2 行
a[:, [1, 3]]     # 取第 1 列和第 3 列
```

```python
# 维度顺序：(行, 列) = (axis=0, axis=1)
# : 出现在哪个位置，就表示"沿着那个轴全取"
a[:, 0]     # 沿行轴全取（所有行），列轴取第 0 → 第一列
a[0, :]     # 行轴取第 0，沿列轴全取 → 第一行
```

> **记忆**：逗号前是行，逗号后是列。`:` = 全要，数字 = 精确位置，列表 = 跳着取。


````python
arr = np.random.randint(1,100,(4,6))
print(arr)
````

````python
print(arr[0,3])
print(arr[:,3])
print(arr[1:3,3])
print(arr[arr>50]) #对二维数组用布尔索引返回一维数组

print(arr[2][arr[2]>50]) #第二行所有大于50的树
print(arr[:,3][arr[:,3]>50]) #第三列所有大于50的数
````
