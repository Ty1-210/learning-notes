## 题目 1：温度数据分析

**涉及**: `mean()`, `max()`, `min()`, 布尔索引


```python
arr = np.array([28,30,29,31,32,30,29])
print('平均气温:%.3f'%np.average(arr))
print('最高气温:', np.max(arr))
print('最低气温:', np.min(arr))
print('气温超过30的天数:', np.sum(np.where(arr>30,1,0)))
```

## 题目 4：随机数据生成

**涉及**: `randint`, 轴向统计, 布尔索引替换


```python
arr = np.random.randint(0,10,(3,4))
print(arr)
print('每列最大值:', np.max(arr,axis=0))
print('每行最小值:', np.min(arr,axis=1))
conditions = [
    arr%2 == 1,
    arr%2 == 0
]
choices = [
    -1, arr
]
arr0 = np.select(conditions, choices) 
#arr0 = np.where(arr%2==1,-1,arr)
#arr[arr%2==-1] = -1
print(arr0)
```

## 题目 5：数组变形

**涉及**: `reshape`, 轴向 sum/mean, `flatten`


```python
arr = np.arange(1,13)
arr = arr.reshape(3,4)
print(arr)
print('每行平均值:',np.mean(arr,axis=1))
print('每列平均值:',np.mean(arr,axis=0))
arr = arr.reshape(-1)
print(arr)
```

## 题目 7：统计函数应用

**涉及**: `sum`, `mean`, `var`, `argmax`, `argmin`


```python
sale = np.array([120,135,110,125,130,140])
print(np.sum(sale))
print(np.mean(sale))
print(np.var(sale))
print(np.argmax(sale)+1)
print(np.argmin(sale)+1)
```

## 题目 8：数组拼接

**涉及**: `hstack`, `vstack`


```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(np.concatenate([[a],[b]],axis=1))
print(np.hstack((a,b)))
print(np.concatenate([[a],[b]],axis=0))
print(np.vstack((a,b)))
```

## 题目 9：唯一值与排序

**涉及**: `unique`, `return_counts`


```python
arr = np.array([2,1,2,3,1,4,3])
u_arr,count = np.unique(arr,return_counts=True)
print(u_arr)
print('元素出现次数:', count)

for i in u_arr:
    print('%d出现次数:'%i, np.sum(np.where(arr==i,1,0)))
```

## 题目 10：综合应用

**涉及**: 利润计算, 派生数组, 统计分析


```python
sales = np.array([20,25,22,30,28])
cost = np.array([15,18,16,22,20])
print(sales-cost)
profit = sales-cost
print(np.mean(profit), np.std(profit))
print(np.argmax(profit)+1)
print(np.where(profit==np.max(profit)))
print(np.argwhere(profit==np.max(profit)))
```
