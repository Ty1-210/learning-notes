## 题目 8：文件读写

1. 把清洗后的 `clean_scores`（题目 4 的结果）保存为 `D:/dsml/data/exam_scores_clean.csv` 并用 `comments` 参数不生成额外注释行。
2. 加载 `matrix_data.npz`，取出 A 和 B 矩阵。
3. 计算 A × B（矩阵乘法）和 A 的逆矩阵。
4. 用 `np.allclose` 验证 A × A_inv 是否约等于单位矩阵。

```python
# ===== 题目 8 答案 =====
# 1. 保存清洗后的成绩
# 沿用题目4的 clean_scores
scores = np.loadtxt(r'D:/dsml/data/exam_scores.csv', delimiter=',', skiprows=1)
clean_scores = scores.copy()
for col in range(clean_scores.shape[1]):
    col_vals = clean_scores[:, col]
    valid = col_vals[(col_vals >= 0) & (col_vals <= 100)]
    col_mean = valid.mean()
    bad = (col_vals < 0) | (col_vals > 100)
    clean_scores[bad, col] = col_mean

np.savetxt(r'D:/dsml/data/exam_scores_clean.csv', clean_scores, delimiter=',', fmt='%.1f',
           header='Math,English,Physics,Chemistry,Biology,CS', comments='')
print('已保存 exam_scores_clean.csv')

# 2/3. 矩阵运算
data = np.load(r'D:/dsml/data/matrix_data.npz')
A, B = data['A'], data['B']
print(f'\nA:\n{A}')
print(f'B:\n{B}')

AB = A @ B
A_inv = np.linalg.inv(A)
print(f'\nA @ B:\n{AB}')
print(f'A 的逆:\n{A_inv}')

# 4. 验证
identity_check = np.allclose(A @ A_inv, np.eye(3))
print(f'\nA @ A_inv 约等于 I? {identity_check}')
```

## 题目 9：综合——股票数据分析

加载 `stock_prices.csv`（跳过表头），完成以下分析：

1. 计算每只股票的日收益率（`diff` 除以前一天收盘价）。
2. 找出收益率超过 3% 的交易日和股票（输出行列索引）。
3. 计算每只股票的 5 日移动平均收盘价。
4. 用 `np.corrcoef` 计算三只股票的相关系数矩阵。
5. 回答：哪两只股票相关性最高？

```python
# ===== 题目 9 答案 =====
stocks = np.loadtxt(r'D:/dsml/data/stock_prices.csv', delimiter=',', skiprows=1)
print(f'股价数据: {stocks.shape}')
print(stocks[:5])

# 1. 日收益率
returns = np.diff(stocks, axis=0) / stocks[:-1]
print(f'\n日收益率前5行:\n{returns[:5]}')

# 2. 收益率 > 3%
big_moves = np.where(returns > 0.03)
print(f'\n收益率 > 3% 的 (行, 股):')
for r, c in zip(big_moves[0], big_moves[1]):
    print(f'  第{r}天, 股票{c}, 收益率={returns[r,c]:.2%}')

# 3. 5日移动平均
ma5 = np.zeros((stocks.shape[0] - 4, stocks.shape[1]))
for i in range(stocks.shape[0] - 4):
    ma5[i] = stocks[i:i+5].mean(axis=0)
print(f'\n5日MA最后3行:\n{ma5[-3:]}')

# 4/5. 相关系数
corr = np.corrcoef(stocks, rowvar=False)
names = ['TechCorp', 'FoodInc', 'SteelCo']
print(f'\n相关系数矩阵:')
for i, n in enumerate(names):
    print(f'  {n}: {corr[i]}')

max_corr = -2
max_pair = (0, 1)
for i in range(3):
    for j in range(i+1, 3):
        if corr[i, j] > max_corr:
            max_corr = corr[i, j]
            max_pair = (i, j)
print(f'\n相关性最高: {names[max_pair[0]]} vs {names[max_pair[1]]}, r={max_corr:.4f}')
```
