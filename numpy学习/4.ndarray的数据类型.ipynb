## ndarray 的数据类型

NumPy 的 `dtype` 比 Python 原生类型丰富得多，每种都有明确的字节宽度。

| 类型 | 说明 | 示例 |
|---|---|---|
| `bool` | 布尔 | True / False |
| `int8/16/32/64` | 有符号整数 | -128~127 (int8) |
| `uint8/16/32/64` | 无符号整数 | 0~255 (uint8) |
| `float16/32/64` | 浮点数 | float64 是默认 |
| `complex64/128` | 复数 | 1+2j |
| `str_` 或 `U` | 字符串 | 固定长度 Unicode |
| `object` | Python 对象 | 存任意 Python 对象 |

```python
a = np.array([1, 2, 3], dtype=np.int32)  # 指定 32 位整数
a.astype(np.float64)                       # 转换类型（不改变原数组）
```

> `float64`（8字节）是默认浮点类型。处理大量数据时，降到 `float32` 可以内存减半且精度损失不大。`astype` 返回新数组，原数组不变。


````python
import numpy as np
````

````python
arr = np.array([1,0,-2,0], dtype='bool') #也可写作 dtype=np.bool
print(arr)
````

````python
arr = np.array([1,0,127,0], dtype='int8') #127是int8上限 128会报错
print(arr)
````
