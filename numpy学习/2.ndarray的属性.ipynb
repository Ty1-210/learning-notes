````python
import numpy as np
````

````python
arr = np.array([[1, 2, 3], [4, 5, 6]]) #两行三列
arr0 = np.array(5)
print(arr)
print('数组的形状:', arr.shape)
print(arr0)
print('数组0的形状:', arr0.shape)
````

````python
print(arr)
print('数组的维度:', arr.ndim)
````

````python
print(arr)
print('元素的个数:', arr.size)
print(arr0)
print('元素的个数:', arr0.size)
````

````python
print('数据类型:', arr.dtype)

arr = np.array([[1, 2.5, 3], [4, 5, 6]])
print(arr)
print('数据类型:', arr.dtype)
````

````python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print('arr的转置;', arr.T)
````

## ndarray 的内存属性

| 属性 | 含义 | 示例 (shape=(2,3), dtype=float64) |
|---|---|---|
| `a.ndim` | 维度数 | 2 |
| `a.shape` | 各维度大小 | (2, 3) |
| `a.size` | 元素总数 | 6 |
| `a.dtype` | 元素类型 | float64 |
| `a.itemsize` | 单个元素字节数 | 8 (float64=8字节) |
| `a.nbytes` | 数组总内存 = size × itemsize | 48 字节 |
| `a.flags` | 内存布局信息（C连续/F连续/只读等） | — |

```python
a = np.array([[1., 2., 3.], [4., 5., 6.]])
a.nbytes    # 6 * 8 = 48
a.strides   # (24, 8) — 跨一行跳 24 字节，跨一列跳 8 字节
```

> `strides` 是理解 NumPy 内存模型的关键——它解释了为什么切片不拷贝数据，只是改了一下"读取跨度"。
