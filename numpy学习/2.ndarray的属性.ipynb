```python
import numpy as np
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]]) #两行三列
arr0 = np.array(5)
print(arr)
print('数组的形状:', arr.shape)
print(arr0)
print('数组0的形状:', arr0.shape)
```

```python
print(arr)
print('数组的维度:', arr.ndim)
```

```python
print(arr)
print('元素的个数:', arr.size)
print(arr0)
print('元素的个数:', arr0.size)
```

```python
print('数据类型:', arr.dtype)

arr = np.array([[1, 2.5, 3], [4, 5, 6]])
print(arr)
print('数据类型:', arr.dtype)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print('arr的转置;', arr.T)
```

itemsize: 单个元素占用的字节内存数
nbytes: 数组总站用内存，size * itemsize
flags: 内存存储方式
