## ✏️ 练习 4-1: 一行两子图

📘 04.08 Multiple Subplots

1 行 2 列, 左 sin 右 cos

**要求:**
- 左图: sin(x) 红色, 右图: cos(x) 蓝色
- 共享 y 轴 (sharey=True)
- figsize=(10, 4)
- 总标题 (fig.suptitle)

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-v0_8-whitegrid')
x = np.linspace(0, 10, 100)
# ---- 开始 ----
fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
ax[0].plot(x, np.sin(x), color='red')
ax[1].plot(x, np.cos(x), color='blue');
# ---- 结束 ----
```

## ✏️ 练习 4-2: 2×2 子图网格

📘 04.08 Multiple Subplots

创建 2×2 子图

**要求:**
- 左上: 折线 y=x² 红色, 右上: 折线 y=x³ 蓝色
- 左下: 散点 y=x²+噪声, 右下: 散点 y=x³+噪声
- x = np.linspace(-3, 3, 50)
- figsize=(10, 8), fig.tight_layout()

```python
x = np.linspace(-3, 3, 50)
# ---- 开始 ----
fig, ax = plt.subplots(2, 2, figsize=(10, 8), sharex=True, sharey=True)
ax[0, 0].plot(x, np.power(x, 2), color='red')
ax[0, 1].plot(x, np.power(x, 3), color='blue')

noise = np.random.normal(0, 0.1, x.shape)      # 均值为0，标准差0.1
ax[1, 0].scatter(x, x + noise, color='red')
ax[1, 1].scatter(x, x + noise, color='blue')
fig.tight_layout();
# ---- 结束 ----
```

## ✏️ 练习 4-3: 不等宽子图 (GridSpec)

📘 04.08 Multiple Subplots

左宽右窄: 股价折线 + 直方图

**要求:**
- 用 plt.GridSpec
- 右图去掉 y 轴标签
- 数据: stock_data.csv (收盘价)

```python
data = np.loadtxt('../data/stock_data.csv', delimiter=',', skiprows=1)
days = data[:, 0]; closing = data[:, 4]
# ---- 开始 ----
grid = plt.GridSpec(1, 7)
grid1 = plt.subplot(grid[0, :5])
grid2 = plt.subplot(grid[0, 5:7], yticklabels=[])

grid1.plot(days, closing)
grid2.hist(closing, bins=len(days));
# ---- 结束 ----
```

## 🌟 练习 4-4: 嵌套子图 (选做)

📘 04.08 Multiple Subplots

大图60天走势, 小图 (fig.add_axes) 放大20-30天

```python
# ---- 开始 ----
# 大图
plt.plot(days, closing)

# 小图（放大镜）：位置手动指定，只画 20-30 天
ax_zoom = plt.axes([0.7, 0.5, 0.3, 0.3])   # [left, bottom, width, height]
ax_zoom.plot(days[20:31], closing[20:31])
ax_zoom.set(xlim=(20, 30))
# ---- 结束 ----
```
