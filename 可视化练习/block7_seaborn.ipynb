## ✏️ 练习 7-1: Seaborn 样式与直方图

📘 04.14 Visualization With Seaborn

使用 Seaborn 绘制分布图

**要求:**
- `sns.set()` 设置 seaborn 风格
- `sns.histplot` 画 500 个正态随机数的直方图, bins=30
- 标题 'Seaborn Histogram'

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

data = np.random.randn(500)
# ---- 开始 ----
sns.set()
sns.histplot(data, bins=30)
plt.title('Seaborn Histogram');
# ---- 结束 ----
```

## ✏️ 练习 7-2: 散点图 + 回归线

📘 04.14 Visualization With Seaborn

用 Seaborn 的 `lmplot` 画带回归线的散点图

**要求:**
- x = randn(100), y = x + 0.5*randn(100) (带噪声的线性关系)
- `sns.lmplot(x=x, y=y)`
- 标题 'Scatter with Regression Line'

```python
# ---- 开始 ----
df = pd.DataFrame({'x': x, 'y': y})
sns.lmplot(x='x', y='y', data=df)
plt.title('Scatter with Regression Line');
# ---- 结束 ----
```

## ✏️ 练习 7-3: 箱线图

📘 04.14 Visualization With Seaborn

用 Seaborn 画箱线图 (boxplot)

**要求:**
- 生成 3 组正态数据, 每组 100 个, 均值分别为 0, 1, 2
- 用列表组合标签, 构成 DataFrame
- `sns.boxplot(x='group', y='value', data=df)`

```python
import pandas as pd
# ---- 开始 ----
groups = ['A']*100 + ['B']*100 + ['C']*100
values = list(np.random.randn(100)) + list(np.random.randn(100)+1) + list(np.random.randn(100)+2)
df = pd.DataFrame({'group': groups, 'value': values})
sns.boxplot(x='group', y='value', data=df)
plt.title('Boxplot by Group');
# ---- 结束 ----
```

## 🌟 练习 7-4: 热力图 (选做)

📘 04.14 Visualization With Seaborn

用 Seaborn 画热力图

**要求:**
- 10×10 随机矩阵
- `sns.heatmap(data, annot=True, cmap='coolwarm')`
- 标题 'Heatmap'

```python
# ---- 开始 ----
data = np.random.randn(10, 10)
sns.heatmap(data, annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Heatmap');
# ---- 结束 ----
```
