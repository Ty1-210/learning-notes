## ✏️ 练习 6-1: 使用内置样式

📘 04.11 Customizing Matplotlib

分别用 'ggplot' 和 'fivethirtyeight' 画 sin/cos

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100); y1 = np.sin(x); y2 = np.cos(x)
# ---- 开始 ----
with plt.style.context('ggplot'):
    fig1, ax1 = plt.subplots()
    ax1.plot(x, y1)
    ax1.plot(x, y2)

with plt.style.context('fivethirtyeight'):
    fig2, ax2 = plt.subplots()
    ax2.plot(x, y1)
    ax2.plot(x, y2)
# ---- 结束 ----
```

## ✏️ 练习 6-2: 自定义 rcParams

📘 04.11 Customizing Matplotlib

通过 plt.rcParams 自定义绘图参数

**要求:**
- 线条宽度: lines.linewidth = 3
- 字体大小: font.size = 14
- 坐标轴标签大小: axes.labelsize = 16
- 画 y=x² 验证效果

```python
# ---- 开始 ----
plt.rc('lines', linewidth=3)
plt.rc('font', size=14)
plt.rc('axes', labelsize=16)
plt.plot(x, np.power(x, 2))
# ---- 结束 ----
```

## ✏️ 练习 6-3: 3D 折线图

📘 04.12 Three-Dimensional Plotting

画出三维螺旋线

**要求:**
- theta = np.linspace(0, 20π, 500)
- x=θcosθ, y=θsinθ, z=θ
- 投影 '3d', 添加坐标轴标签

```python
from mpl_toolkits.mplot3d import Axes3D
# ---- 开始 ----
theta = np.linspace(0, 20*np.pi, 500)
x = theta*np.cos(theta)
y = theta*np.sin(theta)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')
ax.plot(x, y, theta)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('theta')
# ---- 结束 ----
```

## 🌟 练习 6-4: 3D 曲面图 (选做)

📘 04.12 Three-Dimensional Plotting

用 grid_data.csv 画 3D 曲面, cmap='coolwarm', view_init(elev=30, azim=60)

```python
# ---- 开始 ----
import pandas as pd
data = pd.read_csv('../data/grid_data.csv')
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_trisurf(data['x'], data['y'], data['z'], cmap='coolwarm')
ax.view_init(elev=30, azim=60)
# ---- 结束 ----
```

---
## 📌 3D 曲面图：plot_trisurf vs plot_surface

| 函数 | 数据要求 | 适合场景 |
|------|---------|---------|
| `plot_trisurf` | 三列散点 (x, y, z) | 数据不是规则网格，或不想手动 reshape |
| `plot_surface` | 二维网格矩阵 (X, Y, Z) | 数据已经是 meshgrid 格式，或可以 reshape |

**简单口诀：** 三列散点用 trisurf，网格矩阵用 surface
