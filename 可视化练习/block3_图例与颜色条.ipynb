## ✏️ 练习 3-1: 图例位置与样式

📘 04.06 Customizing Plot Legends

画出 sin/cos/sin*cos 并设置图例

**要求:**
- sin: 红实线, cos: 蓝虚线, sin*cos: 绿点划线
- 图例左上角 (loc='upper left')
- 阴影+圆角 (shadow=True, fancybox=True)
- 图例标题 'Functions'

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-v0_8-whitegrid')
x = np.linspace(0, 10, 100)
# ---- 开始 ----
line1 = plt.plot(x, np.sin(x), '-r')
line2 = plt.plot(x, np.cos(x), '--b')
line3 = plt.plot(x, np.sin(x)*np.cos(x), '-.g')
plt.legend(line1+line2+line3, ['sin', 'cos', 'sin*cox'], shadow=True, fancybox=True, loc='upper left', title='Functions', frameon=True)
plt.show();
# ---- 结束 ----
```

## ✏️ 练习 3-2: 散点图 + 颜色条

📘 04.07 Customizing Colorbars

200个随机点, 颜色代表距原点距离

**要求:**
- c=colors, cmap='plasma', alpha=0.6
- 添加颜色条, 标签 'Distance from Origin'
- 颜色条 shrink=0.8

```python
np.random.seed(42)
n = 200
x = np.random.randn(n); y = np.random.randn(n)
colors = np.sqrt(x**2 + y**2)
# ---- 开始 ----
plt.scatter(x, y, c=colors, cmap='plasma', alpha=0.6)
plt.colorbar(shrink=0.8, label='Distance from Origin')
# shrink 控制 colorbar 相对于原图的高度比例，默认是 1.0
# ---- 结束 ----
```

## ✏️ 练习 3-3: 自定义颜色条刻度

📘 04.07 Customizing Colorbars

30×30随机数据, imshow + coolwarm

**要求:**
- 颜色条刻度设为 [-3, -1.5, 0, 1.5, 3]
- 颜色条标签 'Z-score'

```python
data = np.random.randn(30, 30)
# ---- 开始 ----
plt.imshow(data, cmap='coolwarm', origin='lower')
plt.colorbar(ticks=np.arange(-3, 4, 1.5), label='Z-score')
# ---- 结束 ----
```
