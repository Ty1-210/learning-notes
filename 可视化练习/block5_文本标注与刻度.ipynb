## ✏️ 练习 5-1: 添加文本标注

📘 04.09 Text and Annotation

sin 曲线波峰波谷标注

**要求:**
- (π/2, 1) 标注 'Peak', 箭头指向
- (3π/2, -1) 标注 'Valley'
- 原点附近 'y = sin(x)' (plt.text)

💡 *提示: arrowprops=dict(arrowstyle='->')*

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-whitegrid')
x = np.linspace(0, 10, 100); y = np.sin(x)
# ---- 开始 ----
fig, ax = plt.subplots()
ax.plot(x, y)
ax.annotate('Peak', xy=(np.pi/2,1), xytext=(2, -0.5), arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"))
ax.text(3*np.pi/2, -1, 'Valley', color='red', ha='center', va='center')
ax.text(0, 0, 'y = sin(x)', transform=ax.transAxes);
# ---- 结束 ----
```

## ✏️ 练习 5-2: 柱状图 + 数值标签

📘 04.09 Text and Annotation

10个城市人口柱状图并标注

**要求:**
- 柱子 steelblue
- 用 ax.text 在每个柱子顶部标人口数
- 标题 'Top 10 City Populations'

```python
import numpy as np
cities, pops = np.loadtxt('../data/city_population.csv', delimiter=',', skiprows=1,
    dtype={'names': ('city','pop'), 'formats': ('U10','f8')}, unpack=True)
# ---- 开始 ----
fig, ax = plt.subplots()
ax.bar(cities, pops, color='steelblue') # 用于已经分好类，直接画每个类别的值
for city in cities:
    k = np.where(cities == city)[0][0] # 第一个索引取出数组，第二个取出值
    ax.text(k, pops[k], pops[k], ha='center', va='bottom')
# ---- 结束 ----
```

## ✏️ 练习 5-3: 自定义刻度

📘 04.10 Customizing Ticks

股票折线图自定义 x 轴刻度

**要求:**
- 只显示第 1,10,20,30,40,50,60 天
- 刻度标签 f'Day {d}'
- 旋转 x 轴标签 45°

```python
data = np.loadtxt('../data/stock_data.csv', delimiter=',', skiprows=1)
days = data[:, 0].astype(int); close = data[:, 4]
fig, ax = plt.subplots(); ax.plot(days, close)
# ---- 开始 ----
ax.xaxis.set_major_locator(plt.FixedLocator([1,10,20,30,40,50,60]))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x,p: f'Day{x}'))
ax.tick_params(axis='x', rotation=45);
# ---- 结束 ----
```

## 🌟 练习 5-4: 数学公式标注 (选做)

📘 04.09 Text and Annotation

画 f(x)=sin(x²)+cos(x), 用 r'$...$' 标注公式

```python
# ---- 开始 ----
x = np.linspace(0, 5, 100)
y = np.sin(x**2) + np.cos(x)

plt.plot(x, y, label=r'$f(x) = \sin(x^2) + \cos(x)$')
plt.legend(frameon=True);
# ---- 结束 ----
```
