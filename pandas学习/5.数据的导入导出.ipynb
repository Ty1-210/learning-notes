````python
import pandas as pd
````

## CSV 的读写

```python
# 读 — 最常用的导入方式
pd.read_csv('file.csv')
pd.read_csv('file.csv', encoding='utf-8')      # 处理中文
pd.read_csv('file.csv', index_col=0)           # 第 0 列作索引
pd.read_csv('file.csv', parse_dates=['日期'])   # 解析日期列

# 写
df.to_csv('output.csv', index=False)            # 不写行号
df.to_csv('output.csv', encoding='utf-8-sig')   # Excel 友好的编码
```

**关键参数**：`index=False` 几乎必加——谁也不想要多余的行号列。


````python
#数据导入
df = pd.read_csv(r'D:/dsml/data/employees.csv')
print(df)
print(type(df))
print(df.tail())

#数据导出
df = df.tail()
df.to_csv(r'D:/dsml/data/new_employees.csv', index=False) #默认为True 会保存索引
````

## JSON 的读写

```python
# 读 — 数组格式的 JSON 直接读
pd.read_json('data.json')

# 写
df.to_json('output.json', orient='records', indent=2)

# 嵌套 JSON 读不进去？先用 json.load 手动解析
import json
with open('nested.json') as f:
    data = json.load(f)
df = pd.json_normalize(data, record_path='items', meta=['order_id'])
```

| orient | 输出格式 |
|---|---|
| `'records'` | `[{列:值}, ...]` — 数组对象，最常用 |
| `'index'` | `{索引: {列:值}}` |
| `'columns'` | `{列名: {索引:值}}` |
| `'split'` | `{columns, index, data}` |
| `'table'` | 含 schema 的完整表描述 |

**其他格式**：`pd.read_excel()` / `df.to_excel()`、`pd.read_parquet()` / `df.to_parquet()`、`pd.read_sql()` / `df.to_sql()` — 后缀改一下即可，模式完全一样。


````python
df = pd.read_json(r'D:/dsml/data/inventory.json')
print(df)


#read_json是个懒人工具，适合规整的扁平数组
#数据一有嵌套、顶层是对象、或者结构不符合它那几种内置 orient，就得退回到 json.load + pd.DataFrame 的手动路线
import json

with open(r'D:/dsml/data/inventory.json') as f:
    data = json.load(f)

print(data) #字典形式

df = pd.DataFrame(data)
print(df)
print(type(df))
````
