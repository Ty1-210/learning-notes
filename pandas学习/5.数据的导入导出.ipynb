```python
df = pd.read_json(r'D:/dsml/data/inventory.json')
print(df)


#read_json是个懒人工具，适合规整的扁平数组
#数据一有嵌套、顶层是对象、或者结构不符合它那几种内置 orient，就得退回到 json.load + pd.DataFrame 的手动路线
import json

with open(r'D:/dsml/data/inventory.json') as f:
    data = json.load(f)

print(data) #字典形式

df = pd.DataFrame(data)
print(df)
print(type(df))
```
