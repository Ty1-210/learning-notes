## 重复数据处理

### 检测重复值
```python
df.duplicated()           # 每行是否与之前行重复 → 布尔 Series
df.duplicated(subset=['姓名', '科目'])  # 只检查指定列
df.duplicated(keep='first')  # 默认：保留首次出现，后续标为重复
df.duplicated(keep='last')  # 保留最后出现，前面的标为重复
df.duplicated(keep=False)    # 所有重复行都标 True（含首次）
```

### 删除重复值
```python
df.drop_duplicates()                    # 删除重复行，保留第一条
df.drop_duplicates(subset=['姓名'])     # 按姓名去重
df.drop_duplicates(keep='last')        # 保留最后一条
df.drop_duplicates(keep=False)         # 重复的全删，一条不留
```

> `duplicated()` 是侦查，`drop_duplicates()` 是执行。先在 `duplicated().sum()` 看一眼有多少重复，再决定删不删。


````python
import pandas as pd
````

````python
data = {
    'name':['alice','alice','bob','alice','jack','bob'],
    'age':[26,25,30,25,35,30],
    'city':['NY','NY','LA','NY','SF','LA']
}
df = pd.DataFrame(data)
print(df)
````

````python
#检测重复值
print(df.duplicated())
#去重
print(df.drop_duplicates(subset=['name','age'],keep='last')) #根据指定列去重 只有指定列都相同才会去除
#keep决定相同元素保留哪个 默认first保留第一个 last保留最后一个 False表示全部删除
#可以接受axis=1删列 subset必须为不删的维度
````
