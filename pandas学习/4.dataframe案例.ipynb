## 案例：学生成绩分析

综合运用：
- `df.sum(axis=1)` 按行求和（总分）
- `df.mean(axis=1)` 按行求平均
- 布尔索引筛选特定条件
- `sort_values()` 排序 + `head()` 取 Top N


````python
data = {
    '姓名': ['张三', '李四', '王五', '赵六', '钱七'],
    '数学': [85, 92, 78, 88, 95],
    '英语': [90, 88, 85, 92, 80],
    '物理': [75, 80, 88, 85, 90]
}

df = pd.DataFrame(data)
print(df)

#dataframe只要想同时取多列 就必须使用两个中括号 [[]]
df0 = pd.DataFrame(df['姓名'])
df0['平均分'] = df[['数学','英语','物理']].mean(axis=1) #两个中括号使得返回值是dataframe
df0['总分'] = df[['数学','英语','物理']].sum(axis=1)
print(df0)

print(df[(df.数学>90) | (df.英语>85)])

df['总分'] = df[['数学','英语','物理']].sum(axis=1)
print(df.sort_values('总分', ascending=False).head(3)['姓名'].tolist())
print(df.nlargest(3,columns=['总分'])['姓名'].tolist())
````

## 案例：销售数据分析

核心操作：**派生列**（销售额 = 单价 × 销量）、`sort_values()` 排名、`groupby()` 按品类聚合。

```python
df['销售额'] = df['单价'] * df['销量']  # 创建新列
df.groupby('产品')['销售额'].sum()      # 按产品汇总销售额
```


````python
data = {
    '产品名称': ['A', 'B', 'C', 'D'],
    '单价': [100, 150, 200, 120],
    '销量': [50, 30, 20, 40]
}
df = pd.DataFrame(data)

df['销售额'] = df['单价']*df['销量']
print(df)

print(df.产品名称[df.销售额==df.销售额.max()].tolist())
print(df.nlargest(1,columns='销售额')['产品名称'])

print(df.sort_values('销售额',ascending=False))
````

## 案例：电商用户行为分析

典型电商分析流程：
1. 派生总消费金额列
2. `groupby('用户')` 聚合
3. `value_counts()` 看商品热度
4. 布尔索引分群（高消费 vs 低消费）


````python
data = {
    '用户ID': [101, 102, 103, 104, 105],
    '用户名': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    '商品类别': ['电子产品', '服饰', '电子产品', '家居', '服饰'],
    '商品单价': [1200, 300, 800, 150, 200],
    '购买数量': [1, 3, 2, 5, 4]
}
df = pd.DataFrame(data)

df['消费金额'] = df['商品单价']*df['购买数量']
print(df)

print(df.nlargest(1,columns=['消费金额']))

print(round(df.消费金额.mean(),2))

print(df[df['商品类别']=='电子产品'].购买数量.sum())
````
