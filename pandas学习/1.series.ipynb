````python
import pandas as pd
import numpy as np
````

## Series 的创建

`pd.Series()` 可以从列表、字典、标量等创建，是 pandas 最基础的一维数据结构。

```python
pd.Series([1, 2, 3])                    # 从列表创建，索引默认 0,1,2...
pd.Series([1, 2, 3], index=['a','b','c'])  # 自定义索引
pd.Series({'a': 1, 'b': 2})             # 从字典创建，key 变索引
pd.Series(5, index=['x', 'y', 'z'])     # 标量广播，所有值相同
```

**关键点**：索引和数据是绑定的，操作时索引会跟着数据走——这和 NumPy 数组完全不同。


````python
s = pd.Series([1, 2, 3, 4, 5])
print(s)
````

````python
#自定义索引
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s)
#自定义标签
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='kkk')
print(s)
````

````python
#通过字典创建
s = pd.Series({'a':1, 'b':2, 'c':3, 'd':4, 'e':5})
print(s)
#截取部分创建
s1 = pd.Series(s, index=['a', 'b', 'e'])
print(s1)
````

## Series 的属性

| 属性 | 说明 |
|---|---|
| `s.values` | 返回数据部分（NumPy 数组） |
| `s.index` | 返回索引对象 |
| `s.dtype` | 返回数据类型 |
| `s.shape` | 返回形状（一元组） |
| `s.size` | 返回元素总数 |
| `s.name` | Series 的名字（可读可写） |

```python
s = pd.Series([10, 20, 30], index=['a','b','c'], name='数量')
s.values  # → array([10, 20, 30])
s.name    # → '数量'
```


````python
'''
index:Series的索引对象
values:Series的值
dtype或dtypes"Series的元素类型
shape:Series的形状
ndim:Series的维度
size:Series的元素个数
name:Series的名称
loc[]   显式索引，按标签索引或切片 根据实际索引
iloc[]  隐式索引，按位置索引或切片 根据行的位置顺序 0，1，2.....
at[]    使用标签访问单个元素
iat[]   使用位置访问单个元素
'''
````

````python
print(s, '\n')
print(s.index)
print(s.values)
print(s.dtypes)
print(s.shape, s.ndim, s.size, s.name)
print('\n')
print(s.loc['a'])
print(s.loc['a':'c']) # loc,iloc 传入切片时返回有 index
print('\n')
print(s.iloc[0])
print(s.iloc[0:2])
print('\n')
print(s.at['a']) #不支持切片
print(s.iat[3])
````

## Series 的访问方式

### 按标签索引 `.loc[]`
```python
s.loc['a']       # 单个标签
s.loc['a':'c']   # 标签切片（含头含尾！和 Python 切片不同）
```

### 按位置索引 `.iloc[]`
```python
s.iloc[0]        # 第 0 个
s.iloc[0:2]      # 位置切片（含头不含尾，和 Python 一致）
```

### 布尔索引
```python
s[s > 15]        # 筛选大于 15 的值
```

> **注意**：`loc` 切片**包含**右端点（标签），`iloc` 切片**不包含**右端点（位置）。这是初学最容易踩的坑。


````python
print(s['c']) #直接访问
print(s[s<3]) #布尔索引
s['f'] = 6
print(s.head()) #默认前五行
print(s.head(3))
print(s.tail()) #默认后五行
print(s.tail(6))
````

## Series 的常用方法

| 方法 | 作用 |
|---|---|
| `describe()` | 统计摘要（count / mean / std / min / 25% / 50% / 75% / max） |
| `value_counts()` | 每个值出现的频数 |
| `sort_values()` | 按值排序 |
| `sort_index()` | 按索引排序 |
| `head(n)` / `tail(n)` | 看前/后 n 行 |
| `unique()` / `nunique()` | 去重值 / 去重后个数 |
| `isin([...])` | 判断值是否在列表中，返回布尔 Series |
| `between(a, b)` | 判断值是否在 [a, b] 区间内 |

```python
s.describe()           # 一眼看清分布
s.value_counts()       # 最常用的探索性操作
s.sort_values(ascending=False)  # 从大到小
```


````python
s =pd.Series([10,2,np.nan,None,4,5,6],index=['a','b','c','d','e','f','g'],name='Data')
print(s)
````

````python
s.head(3)
s.tail(4)
````

````python
#查看所有描述性信息
s.describe() #count会默认去掉缺失值
````

````python
print(s.count())
print(s.keys()) #用方法获得索引
print(s.index) #用属性获得索引
````

````python
#判断缺失值
print(s.isna()) 
s.isna()
````

````python
#判断series值是否在集合中
s.isin([4,5,6])
````

````python
#统计方法 一般都省略NaN
print(s.mean())
print(s.std())
print(s.median())
print(s.var())
print(s.min())
print(s.max())
print(s.quantile(0.3)) #4+1*(4*0.3-1) 4个间隔
````

````python
print(s.mode())
s['h'] = 10
print(s.mode())
````

````python
#排序
print(s.sort_values()) #缺失值放在最后
print(s.sort_values(ascending=False))
print(s.sort_index()) #按索引排序
print(s.sort_index(ascending=False))
````

````python
#计数
s.value_counts()
````

````python
#去重
print(s.nunique()) #去重后元素个数
print(s)
print(s.drop_duplicates()) #效果等同于s.unique() 一个返回Series 一个返回array
print(s)
print(s.drop_duplicates(inplace=True)) #inplace控制是否在原数据中修改
print(s)
````
