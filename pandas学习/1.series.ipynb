```python
import pandas as pd
import numpy as np
```

创建

```python
s = pd.Series([1, 2, 3, 4, 5])
print(s)
```

```python
#自定义索引
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s)
#自定义标签
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='kkk')
print(s)
```

```python
#通过字典创建
s = pd.Series({'a':1, 'b':2, 'c':3, 'd':4, 'e':5})
print(s)
#截取部分创建
s1 = pd.Series(s, index=['a', 'b', 'e'])
print(s1)
```

属性

```python
'''
index:Series的索引对象
values:Series的值
dtype或dtypes"Series的元素类型
shape:Series的形状
ndim:Series的维度
size:Series的元素个数
name:Series的名称
loc[]显式索引，按标签索引或切片
iloc[]  |隐式索引，按位置索引或切片
at[]    使用标签访问单个元素
iat[]   使用位置访问单个元素
'''
```

```python
print(s, '\n')
print(s.index)
print(s.values)
print(s.dtypes)
print(s.shape, s.ndim, s.size, s.name)
print('\n')
print(s.loc['a'])
print(s.loc['a':'c'])
print('\n')
print(s.iloc[0])
print(s.iloc[0:2])
print('\n')
print(s.at['a']) #不支持切片
print(s.iat[3])
```

访问数据

```python
print(s['c']) #直接访问
print(s[s<3]) #布尔索引
s['f'] = 6
print(s.head()) #默认前五行
print(s.head(3))
print(s.tail()) #默认后五行
print(s.tail(6))
```

常用方法

```python
s =pd.Series([10,2,np.nan,None,4,5,6],index=['a','b','c','d','e','f','g'],name='Data')
print(s)
```

```python
s.head(3)
s.tail(4)
```

```python
#查看所有描述性信息
s.describe() #count会默认去掉缺失值
```

```python
print(s.count())
print(s.keys()) #用方法获得索引
print(s.index) #用属性获得索引
```

```python
#判断缺失值
print(s.isna()) 
s.isna()
```

```python
#判断series值是否在集合中
s.isin([4,5,6])
```

```python
#统计方法 一般都省略NaN
print(s.mean())
print(s.std())
print(s.median())
print(s.var())
print(s.min())
print(s.max())
print(s.quantile(0.3)) #4+1*(4*0.3-1) 4个间隔
```

```python
print(s.mode())
s['h'] = 10
print(s.mode())
```

```python
#排序
print(s.sort_values()) #缺失值放在最后
print(s.sort_values(ascending=False))
print(s.sort_index()) #按索引排序
print(s.sort_index(ascending=False))
```

```python
#计数
s.value_counts()
```

```python
#去重
print(s.nunique()) #去重后元素个数
print(s)
print(s.drop_duplicates()) #效果等同于s.unique() 一个返回Series 一个返回array
print(s)
print(s.drop_duplicates(inplace=True)) #inplace控制是否在原数据中修改
print(s)
```
