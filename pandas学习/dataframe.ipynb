```python
import pandas as pd
import numpy as np
```

创建

```python
#通过series创建
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,10])
df = pd.DataFrame({'第一列':s1, '第二列':s2})
df.index = [1010,1011,1012,1013,1014] #不能在创建的时候改index dataframe和series索引必须对对齐
print(df)
print(type(df))
print(type(df['第一列']))
```

```python
#通过字典创建
df = pd.DataFrame(
    {
        'ID':[1,2,3,4,5],
        'NAME':['A','B','C','D','E'],
        'AGE':[12,13,21,20,12]
    },
    index=[1010,1011,1012,1013,1014],
    columns=['ID','AGE','NAME'] #既可以改变列顺序，也可以达成对列显示的筛选
)
print(df)
```

属性

```python
print('行索引:',df.index)
print('列标签:',df.columns)
print('值:',df.values)
```

```python
print('维度:',df.ndim)
print('数据类型:',df.dtypes)
print('形状:',df.shape)
print('元素个数:',df.size)
```

```python
print('转置:',df.T)
dft = df.T
print('行索引:',dft.index)
print('列标签:',dft.columns)
print('值:',dft.values)
```

```python
print(df)
print(df.loc[1011:1014])
print(df.iloc[1:5])
#如果索引恰好是 0, 1, 2, 3...，那 loc 和 iloc 看起来确实一样。但一旦索引被打乱就不同
#e.g. 如果我的索引是1，2，4，3，5，loc[2:3]是2，4，3；iloc[2:3]是4
#loc左闭右闭 iloc左闭右开

print(df.loc[1012:1014,'AGE':'NAME']) #dataframe中可获取列，但先必须规定行
print(df.iloc[2:5,1:3])
```

```python
print(df)
print(df.at[1010,'NAME'])
print(df.iat[0,2])
```
