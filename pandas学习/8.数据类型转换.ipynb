## 数据类型转换

### 查看类型
```python
df.dtypes          # 每列的数据类型
df['列名'].dtype   # 单列类型
df.info()          # 完整信息（含非空数、内存）
```

### 转换类型
```python
# astype — 通用转换
df['列'] = df['列'].astype('int')       # 转整数
df['列'] = df['列'].astype('float')     # 转浮点
df['列'] = df['列'].astype('string')    # 转 string（pandas 1.0+）
df['列'] = df['列'].astype('category')  # 转分类

# 专用转换
pd.to_numeric(df['列'], errors='coerce')  # 强转数字，失败变 NaN
pd.to_datetime(df['列'])                  # 转日期时间
```

### object vs string vs category
| 类型 | 用途 | 内存 |
|---|---|---|
| `object` | 混装类型（字符串/数字/列表），通用但效率低 | 高 |
| `string` | 纯字符串列，字符串操作更快 | 中 |
| `category` | 重复值多的列（如部门、品类、地区），底层存整数编码 | 最低 |

### category 专有操作
```python
df['列'].cat.categories        # 所有合法类别
df['列'].cat.codes             # 底层整数编码
df['列'].cat.rename_categories({'旧': '新'})
df['列'].cat.add_categories(['新类别'])
df['列'].cat.remove_unused_categories()
```

> **口诀**：重复值多且有限 → `category`；纯文本 → `string`；类型混杂 → `object`。`astype` 默认不改变原 df，要 `df['列'] = df['列'].astype(...)` 赋值回去。


````python
import pandas as pd
````

````python
df = pd.read_csv(r'D:\dsml\data\products.csv')
print(df)
````

````python
#数值类型转换
print(df.dtypes)
df['库存'] = df['库存'].astype('int16')
print(df.dtypes)
df['成本价'] = df['成本价'].round(1)
print(df)
````

````python
#
print(df.品类.drop_duplicates())
df['品类'] = df['品类'].astype('category') #省内存、加速操作。不是把每行存一遍 而是内部建一个映射表
#.cat.codes自动把category转化为类似索引的数字编号
print(df['品类'].cat.codes)
#.cat.categories显示这个 category 列里所有合法的类别值，返回一个 Index
print(df['品类'].cat.categories)
print(df.dtypes)
````
