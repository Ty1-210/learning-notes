## DataFrame 的创建

`pd.DataFrame()` 可以从字典、列表、外部文件等多种来源创建。

```python
# 从字典创建 — 最常用方式，key 变列名，value 变列数据
pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]})

# 从嵌套列表创建，手动指定列名
pd.DataFrame([[1,4],[2,5],[3,6]], columns=['A','B'])

# 从 CSV 读取
pd.read_csv('file.csv')

# 从 JSON 读取
pd.read_json('file.json')
```

**字典 vs 列表**：字典的 key 自动成列名，更简洁；列表要额外传 `columns`。


````python
#通过series创建
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,10])
df = pd.DataFrame({'第一列':s1, '第二列':s2})
df.index = [1010,1011,1012,1013,1014] #不能在创建的时候改index dataframe和series索引必须对对齐
print(df)
print(type(df))
print(type(df['第一列']))
````

````python
#通过字典创建
df = pd.DataFrame(
    {
        'ID':[1,2,3,4,5],
        'NAME':['A','B','C','D','E'],
        'AGE':[12,13,21,20,12]
    },
    index=[1010,1011,1012,1013,1014],
    columns=['ID','AGE','NAME'] #既可以改变列顺序，也可以达成对列显示的筛选
)
print(df)
````

## DataFrame 的属性

| 属性 | 说明 |
|---|---|
| `df.shape` | (行数, 列数) |
| `df.columns` | 列名索引 |
| `df.index` | 行索引 |
| `df.dtypes` | 每列的数据类型 |
| `df.values` | 底层 NumPy 数组（类型不一致时会是 object） |
| `df.info()` | 结构化摘要：列名、非空数、dtype、内存占用 |

```python
df.shape       # → (100, 5) — 100 行 5 列
df.info()      # 进新数据必跑的第一条
df.describe()  # 数值列的统计摘要（count/mean/std/min/四分位/max）
```

> `info()` 看结构，`describe()` 看分布。两条命令是新数据到手后的标准起手式。

## DataFrame 的访问方式

### 按列名
```python
df['列名']          # 单列 → Series
df[['A', 'B']]      # 多列 → DataFrame（注意双层括号）
```

### 按行列位置 `.iloc[]`
```python
df.iloc[0, 1]       # 第 0 行第 1 列
df.iloc[0:3, :]     # 前 3 行所有列
df.iloc[:, [0, 2]]  # 所有行的第 0、2 列
```

### 按行列标签 `.loc[]`
```python
df.loc[2, '姓名']             # 标签定位
df.loc[0:4, ['姓名','成绩']]   # 标签切片（含尾！）
```

### 布尔索引
```python
df[df['成绩'] > 90]                          # 单条件
df[(df['成绩'] > 90) & (df['科目'] == '数学')] # 多条件用 & | ~
```

> **核心区分**：`.loc` 用标签（含头含尾），`.iloc` 用位置（含头不含尾），`df['列名']` 直接取列。


````python
#获取单列数据
print(df['NAME'])
print(type(df['NAME']))
print(df.NAME)
print(type(df.NAME))
print(df[['NAME']])
print(type(df[['NAME']])) #这种写法的数据类型是dataframe
````

````python
#查看部分数据
print(df.tail())
print(df.tail(2))
print(df.head())
print(df.head(3))
````

````python
#布尔索引
df[(df.AGE>15) & (df.ID>2)]
````

````python
#随机抽样
print(df.sample())
print(df.sample(3))
````

## DataFrame 的常用方法

| 方法 | 作用 |
|---|---|
| `head(n)` / `tail(n)` | 前/后 n 行 |
| `info()` / `describe()` | 结构摘要 / 数值统计 |
| `sort_values(by='列名')` | 按指定列排序 |
| `value_counts()` | 列值频数统计 |
| `drop(labels, axis=1)` | 删除列 (axis=1) 或行 (axis=0) |
| `drop_duplicates()` | 删除重复行 |
| `rename(columns={...})` | 重命名列 |
| `set_index()` / `reset_index()` | 设置/重置索引 |
| `isnull()` / `notnull()` | 缺失值检测 |
| `fillna()` / `dropna()` | 缺失值处理 |
| `groupby()` | 分组聚合 |
| `merge()` / `concat()` | 合并拼接 |

```python
df.sort_values('成绩', ascending=False).head(3)  # 前三名
df['科目'].value_counts()                        # 每科出现次数
```


````python
print(df.isin([4,5,'E'])) #查看元素是否在参数集合中
print(df.isna()) #查看是否有缺失值
````

````python
print(df.sum())
print(df['ID'].sum())
print(df.groupby('ID').sum()) #相同id的不同行按列一一求和
print(df.AGE.max())
print(df.AGE.min())
print(df.AGE.mean())
print(df.AGE.median())
print(df.AGE.quantile(0.75))
print(df.AGE.quantile(0.25))
print(df.AGE.mode())
print(df.AGE.std())
print(df.AGE.var())
````

````python
print(df.describe())
print(df.value_counts()) #两行所有元素必须对应相等才能视为重复值
````

````python
print(df)
print(df.drop_duplicates(subset='AGE')) #去重
print(df.duplicated(subset='AGE')) #检查是否有重复 只有在第二次出现的行返回True 第一次出现视为'原版' 返回False
````

````python
print(df.replace(13,20)) #不关心列 所有指定值都会被替换 不改变原dataframe除非inplace=True
print(df['AGE'].replace(13,20))
````

````python
#累计
print(df.cumsum())
print(df.cummax())
print(df.cummin())
#默认按列 传入axis=1可以按行累计
````

````python
#排序
print(df.sort_index(ascending=False))
print(df.sort_values(by=['AGE','NAME'], ascending=[False,True]))
````

````python
print(df.nlargest(2,columns=['AGE']))
print(df.nsmallest(1,columns=['AGE'])) #不能接收数字意外的元素 不会按照字符ASCII来排序
````
