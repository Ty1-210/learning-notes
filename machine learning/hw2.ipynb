```python
import torch
import numpy as np
```

```python
tensor = torch.rand(5, 8)
print(tensor)
# 索引第2至3行
rows = tensor[1:3, :] #Python的索引从0开始，所以第2至3行是索引1:3
print(rows)
# 索引奇数列
odd = tensor[:, 1::2]
print(odd)
# 索引所有>0.6的元素
elements = tensor[tensor > 0.6]
print(elements)
```

```python
x = torch.rand(1, 5)
print(x)
y = torch.rand(6, 1)
print(y)
# 按照广播机制相加
s = x + y
print(s)
t = x.expand(6, 5) + y.expand(6, 5) #需要x和y的形状相同，因此需要扩展为6*5
print(t)
# 将广播相加的结果赋值给x
x = t
print(x)
```

```python
a1 = torch.rand(8, 8)
b1 = torch.rand(8, 8)
print(a1)
print(b1)
# 将张量转化为numpy的array
a2 = a1.numpy()
b2 = b1.numpy()
print(a2)
print(b2)
# 做矩阵相乘
result1 = np.dot(a2, b2)
print(result1)
# 将结果转化为pytorch下的张量类型
result2 = torch.from_numpy(result1)
print(result2)
```
