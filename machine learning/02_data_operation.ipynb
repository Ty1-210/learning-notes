```python
indices = torch.LongTensor([0,2]) # Long
print(indices) 
print(indices.dtype) 
z = torch.index_select(x,0,indices) # 从 x 里按 indices 指定的行号取出行，存到 z
print(z) 
z = torch.index_select(x,1,indices) 
print(z) 
z += 1 
print(x) 
print(z) 
#0指行，1指列 
```

```python
x = torch.randn(5,3,dtype=torch.float)
print(x) 
mask = x.ge(0.5) # 对 x 里每个元素判断是否大于等于 0.5，返回一个同形状的布尔 tensor
print(mask) 
print(torch.masked_select(x,mask))
# masked_select 按布尔 tensor 取值，等价于 x[mask]
y = x[x>0.5] 
print(y == x[mask])
y += 1 
print(y)
#张量 mask须跟input张量有相同数量的元素数目，但形状或维度不需要相同。 
#注意： 返回的张量不与原始张量共享内存空间。
```

```python
print(x) 
y = x.view(15)  # view 是 PyTorch 的重塑形状，和 NumPy 的 reshape 一样
z = x.view(-1,5) 
print(y) 
print(z) 
print(y.shape,z.shape) 
print(x) 
x += 1 
print(x) 
print(y) 
y += 1 
print(x) 
print(y) 
#view()返回的新tensor和源tensor共享内存 
#列顺序
```

```python
x = torch.randn(4,6) 
print(x) 
y = x.view(-1,2) 
print(y) 
y = x.view(4,-1) 
print(y) 
y = x.view(3,8) 
print(y) 
# -1表示对该维的数目不做限制
```

```python
x = torch.randn([3, 5])
x_cp = x.clone().view(15)
print(x_cp)
x += 1 
print(x) 
print(x_cp) 
x_cp += 3 
print(x) 
print(x_cp) 
#克隆后的不共享内存
```

```python
x = torch.tensor([1,2]) 
y = torch.tensor([3,4]) 
id_before = id(y) # id 返回对象的内存地址
y = y + x 
print(id(y) == id_before) 
print(id(y)) 
print(id_before) 
```

```python
x = torch.tensor([1,2]) 
y = torch.tensor([3,4]) 
id_before = id(y) 
y[:] = y + x 
print(id(y) == id_before) 
print(id(y)) 
print(id_before)
```

```python
x = torch.tensor([1,2])
y = torch.tensor([3,4]) 
id_before = id(y) 
torch.add(x,y,out=y) #或y += x和y.add_(x) 
print(id(y) == id_before) 
```

```python
#Tensor转Numpy，共享内存 
a = torch.ones(5) 
print(a) 
b = a.numpy() 
print(a,b) 
a += 1 
print(a,b) 
b += 1 
print(a,b) 
```

```python
#Numpy转Tensor，共享内存 
a = numpy.ones(5) 
b = torch.from_numpy(a) 
print(a,b) 
a += 1 
print(a,b) 
b += 1 
print(a,b) 
```

```python
c = torch.tensor(a) 
a += 1 
print(a,c) 
#这种办法不共享内存 
```
